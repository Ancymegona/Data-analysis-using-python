## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
data = pd.read_csv("/content/Dataset .csv")

print("Dataset Preview:")
print(data.head())

positive_words = ["good", "great", "amazing", "excellent", "awesome", "nice", "loved", "friendly", "delicious"]
negative_words = ["bad", "worst", "poor", "terrible", "rude", "slow"]

positive_count = Counter()
negative_count = Counter()

review_lengths = []
ratings = []

for review, rating in zip(data["Rating text"], data["Aggregate rating"]):
    # Ensure review is a string before calling .lower() and .split()
    if isinstance(review, str):
        words = review.lower().split()
        review_lengths.append(len(words))
    else:
        review_lengths.append(0) # Handle non-string reviews

    ratings.append(rating)

    if isinstance(review, str):
        for word in words:
            if word in positive_words:
                positive_count[word] += 1
            if word in negative_words:
                negative_count[word] += 1

print("\nMost Common Positive Keywords:")
print(positive_count.most_common())

print("\nMost Common Negative Keywords:")
print(negative_count.most_common())

# Check if review_lengths is empty to avoid division by zero
if review_lengths:
    average_length = sum(review_lengths) / len(review_lengths)
    print("\nAverage Review Length:", round(average_length, 2), "words")
else:
    print("\nNo reviews processed, cannot calculate average length.")

# Check if there are any reviews to plot
if review_lengths and ratings:
    plt.scatter(review_lengths, ratings)
    plt.xlabel("Review Length (Number of Words)")
    plt.ylabel("Rating")
    plt.title("Review Length vs Rating")
    plt.show()
else:
    print("No data to plot: review_lengths or ratings are empty.")
````

````python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("/content/Dataset .csv")

print("\nDataset Preview:")
print(data.head())

highest_votes = data.loc[data["Votes"].idxmax()]
print("\nRestaurant with Highest Votes:")
print(highest_votes)

lowest_votes = data.loc[data["Votes"].idxmin()]
print("\nRestaurant with Lowest Votes:")
print(lowest_votes)

correlation = data["Votes"].corr(data["Aggregate rating"])
print("\nCorrelation between Votes and Rating:", round(correlation, 3))

plt.scatter(data["Votes"], data["Aggregate rating"])
plt.xlabel("Number of Votes")
plt.ylabel("Rating")
plt.title("Votes vs Rating Relationship")
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("/content/Dataset .csv")

print("\nDataset Preview:")
print(data.head())


data["Has Online delivery"] = data["Has Online delivery"].map({"Yes": 1, "No": 0})
data["Has Table booking"] = data["Has Table booking"].map({"Yes": 1, "No": 0})


delivery_by_price = data.groupby("Price range")["Has Online delivery"].mean() * 100
booking_by_price = data.groupby("Price range")["Has Table booking"].mean() * 100


print("\nOnline Delivery availability by Price Range (%):")
print(delivery_by_price)

print("\nTable Booking availability by Price Range (%):")
print(booking_by_price)

delivery_by_price.plot(kind="bar", title="Online Delivery vs Price Range", ylabel="Percentage")
plt.show()

booking_by_price.plot(kind="bar", title="Table Booking vs Price Range", ylabel="Percentage")
plt.show()

delivery_corr = data["Price range"].corr(data["Has Online delivery"])
booking_corr = data["Price range"].corr(data["Has Table booking"])

print("\nCorrelation between Price Range and Online Delivery:", round(delivery_corr, 3))
print("Correlation between Price Range and Table Booking:", round(booking_corr, 3))
````
